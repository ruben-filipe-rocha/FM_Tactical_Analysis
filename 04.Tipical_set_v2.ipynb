```python
# ==========================
# 1. Imports + Paths + Leitura
# ==========================
import numpy as np
import pandas as pd

ARQ_IN = "Matrix.csv"  # input
X = pd.read_csv(ARQ_IN, index_col=0).clip(lower=0).astype(float)
X  # mostra
```

```python
# ==========================
# 2. Typical Set (Shannon)
# ==========================
P = X.div(X.sum(axis=1), axis=0).fillna(0)
H_frame = -np.sum(P * np.log2(P + 1e-12), axis=1)
H_mean, H_std = H_frame.mean(), H_frame.std()
lower, upper = H_mean - H_std, H_mean + H_std
mask_typical = (H_frame >= lower) & (H_frame <= upper)
X_typical = X[mask_typical]
X_discarded = X[~mask_typical]
X_typical.head(10)
```

```python
# ==========================
# 3. Resumo (total, tpicos, descartados)
# ==========================
total_frames = len(X)
typical_frames = int(mask_typical.sum())
discarded_frames = total_frames - typical_frames

df_resumo = pd.DataFrame({
    'Total_frames':[total_frames],
    'Frames_tpicos':[typical_frames],
    'Descartados':[discarded_frames],
    'Percentagem_tpicos (%)':[round(100*typical_frames/total_frames,2)]
})
df_resumo
```

```python
# ==========================
# 4. Descritivo descartados (timestamp + clulas ativas)
# ==========================
num_celulas_ativas_desc = (X_discarded > 0).sum(axis=1)
df_descartados = pd.DataFrame({
    'timestamp': X_discarded.index,
    'celulas_ativas': num_celulas_ativas_desc
}).reset_index(drop=True).sort_values('celulas_ativas', ascending=False)
df_descartados.head(15)
```

```python
# ==========================
# 5. Histogramas discretos (tpicos vs descartados)
# ==========================
contagens_tipicos = (X_typical > 0).sum(axis=1)
resumo_tipicos = contagens_tipicos.value_counts().sort_index().reset_index()
resumo_tipicos.columns = ['n_celulas_ativas','n_timestamps']
resumo_tipicos
```

```python
# ==========================
# 6. Histogramas descartados
# ==========================
contagens_desc = (X_discarded > 0).sum(axis=1)
resumo_descartados = contagens_desc.value_counts().sort_index().reset_index()
resumo_descartados.columns = ['n_celulas_ativas','n_timestamps']
resumo_descartados
```

```python
# ==========================
# 7. Output final 
# ==========================
X_typical.to_csv("Matrix_X_Typical_set.csv", index=True)
#X_discarded.to_csv("frames_atipicos.csv", index=True)

```
