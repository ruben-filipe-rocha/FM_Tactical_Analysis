```python
'''
# ==========================
# 1) Polars: equipa do actor por frame + normalização de coordenadas
# ==========================
import polars as pl

# ==========================
# 1.1) Constantes do campo
# ==========================
FIELD_LENGTH = 120.0  # metros
FIELD_WIDTH  = 80.0   # metros (usar 60.0 se for 120x60)

# ==========================
# 1.2) Derivar equipa do actor por (period, timestamp)
# ==========================
actor_equipa = (
    df
    .filter(pl.col("actor") == True)
    .select(["timestamp", "period", "team"])
    .unique(subset=["timestamp", "period"], keep="first")
    .rename({"team": "equipa_actor_frame"})
)

# ==========================
# 1.3) Juntar essa info ao df
# ==========================
df = df.join(actor_equipa, on=["timestamp", "period"], how="left")

# ==========================
# 1.4) Equipas únicas e equipa do actor no 1.º frame (minuto 0, período 1)
# ==========================
equipas_unicas = (
    df.select(pl.col("team").drop_nulls().unique())
      .to_series(0)
      .to_list()
)

equipa_actor_inicial = (
    df.filter((pl.col("period") == 1) & (pl.col("minute") == 0) & (pl.col("actor") == True))
      .sort("timestamp")
      .select(pl.col("team").first())
      .to_series()
      .item()
)

equipa_oponente = next(e for e in equipas_unicas if e != equipa_actor_inicial)

# ==========================
# 1.5) Assumir orientação: actor inicial à esquerda
# ==========================
equipa_esquerda = equipa_actor_inicial
equipa_direita  = equipa_oponente

# ==========================
# 1.6) Normalizar coordenadas (x_norm, y_norm)
# ==========================
cond_inverte = (
    pl.col("equipa_actor_frame").is_not_null()
    & (pl.col("equipa_actor_frame") != pl.lit(equipa_esquerda))
)

df = df.with_columns([
    pl.when(cond_inverte).then(FIELD_LENGTH - pl.col("x")).otherwise(pl.col("x")).alias("x_norm"),
    pl.when(cond_inverte).then(FIELD_WIDTH  - pl.col("y")).otherwise(pl.col("y")).alias("y_norm"),
])

# ==========================
# 1.7) (Opcional) Seleção rápida para inspeção
# ==========================
# df.select(["period","timestamp","team","equipa_actor_frame","x","y","x_norm","y_norm"]).head()
'''
```

```python
# ==========================
# Imports
# ==========================
import polars as pl

# ==========================
# Função de normalização (Polars)
# ==========================
def normalizar_coordenadas(
    df: pl.DataFrame | pl.LazyFrame,
    equipa_esquerda: str,
    field_length: float = 120.0,
    field_width: float = 80.0
) -> pl.DataFrame:
    # se for LazyFrame, recolher
    if isinstance(df, pl.LazyFrame):
        df = df.collect()

    # garantir que x e y são numéricos
    df = df.with_columns([
        pl.col("x").cast(pl.Float64, strict=False),
        pl.col("y").cast(pl.Float64, strict=False)
    ])

    # condição de inversão: equipa_actor_frame não nulo e diferente de equipa_esquerda
    mask = (pl.col("equipa_actor_frame").is_not_null()) & (pl.col("equipa_actor_frame") != equipa_esquerda)

    # adicionar colunas x_norm e y_norm
    df = df.with_columns([
        pl.when(mask)
          .then(field_length - pl.col("x"))
          .otherwise(pl.col("x"))
          .alias("x_norm"),
        pl.when(mask)
          .then(field_width - pl.col("y"))
          .otherwise(pl.col("y"))
          .alias("y_norm")
    ])

    return df


equipa_esquerda = "NomeDaEquipaEsquerda"  # define a equipa que ataca da esquerda

df = normalizar_coordenadas(df, equipa_esquerda)

# agora df inclui as colunas x_norm e y_norm
df.select(["x", "y", "x_norm", "y_norm"]).head(5)
```

```python
# ==========================
# 1. Imports e parâmetros
# ==========================
from shapely.geometry import Point
from shapely.geometry import Point, Polygon

from grid_hex import gerar_hex_grid  # usa a mesma função da animação
import pandas as pd
import numpy as np

# parâmetros do campo e grelha
FIELD_WIDTH = 120
FIELD_HEIGHT = 80
HEX_SIZE = 1.86
X0, Y0 = 0.0, 0.0

# ==========================
# 2. Geração da grelha hexagonal
# ==========================
cell_coords = gerar_hex_grid(  # <--- guarda aqui as coordenadas
    field_width=FIELD_WIDTH,
    field_height=FIELD_HEIGHT,
    hex_size=HEX_SIZE
)

cell_polygons = [Polygon(coords) for coords in cell_coords]  # converte listas em Polygons
cell_centers = [poly.centroid.coords[0] for poly in cell_polygons]
n_cells = len(cell_polygons)

# ==========================
# 3. Função: processar 1 frame
# ==========================
def processar_frame(frame_data):
    grid_A = [0] * n_cells  # nº de jogadores da equipa A por célula

    df_azuis = frame_data[frame_data["teammate"] == True].dropna(subset=["x_norm", "y_norm"])

    for _, row in df_azuis.iterrows():
        p = Point(row["x_norm"], row["y_norm"])
        for i, poly in enumerate(cell_polygons):
            if poly.contains(p):
                grid_A[i] = 1  # célula marcada como ocupada (não acumula)
                break

    df_resultado = pd.DataFrame({
        "cell_index": range(n_cells),
        "centro_x": [c[0] for c in cell_centers],
        "centro_y": [c[1] for c in cell_centers],
        "jogadores_A": grid_A
    })
    return df_resultado


# ==========================
# 4. Aplicar a todos os frames do jogo
# ==========================
frames_resultado = []

# garante que df contém: timestamp, period, x_norm, y_norm, teammate, actor
frames_unicos = df[["timestamp", "period"]].drop_duplicates()

for _, row in frames_unicos.iterrows():
    ts, period = row["timestamp"], row["period"]
    frame_data = df[(df["timestamp"] == ts) & (df["period"] == period)]
    df_frame = processar_frame(frame_data)
    df_frame["timestamp"] = ts
    df_frame["period"] = period
    frames_resultado.append(df_frame)

df_resultado_final = pd.concat(frames_resultado, ignore_index=True)

# ==========================
# 5. Construção da matriz final
# ==========================
df_resultado_final["jogadores_A"] = pd.to_numeric(df_resultado_final["jogadores_A"], errors="coerce").fillna(0).astype(int)

X_A = (
    df_resultado_final
      .pivot_table(index="timestamp", columns="cell_index", values="jogadores_A", aggfunc="max")
      .sort_index()
      .fillna(0)
)


X_A.columns = X_A.columns.astype(int)
X_A = X_A.reindex(sorted(X_A.columns), axis=1)

# ==========================
# 6. Guardar resultado
# ==========================
# caminho relativo à pasta notebooks → Files
X_A.to_csv("../Files/X_A_grid.csv", index=True)
# alternativa mais rápida:
# X_A.to_parquet("../Files/X_A_grid.parquet")

# ==========================
# 7. Verificação
# ==========================
X_A.head()
```

```python
# ==========================
# 8. Validação rápida
# ==========================
for ts, row in X_A.iterrows():
    # seleciona apenas colunas com valor > 0
    ativos = row[row > 0]
    if not ativos.empty:
        print(f"\nTimestamp: {ts}")
        print(ativos)
```
